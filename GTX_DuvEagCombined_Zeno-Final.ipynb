```python
XD.describe()
```

```python
XD.info()
```

```python
XD["kb_elev_m"]= XD["kb_elev_m"].fillna(XD["kb_elev_m"].median())
XD["pressure_recorder_depth_m"]= XD["pressure_recorder_depth_m"].fillna(XD["pressure_recorder_depth_m"].median())
XD["initial_hydrostatic_pressure_kpa"]= XD["initial_hydrostatic_pressure_kpa"].fillna(XD["initial_hydrostatic_pressure_kpa"].median())
XD["final_hydrostatic_pressure_kpa"]= XD["final_hydrostatic_pressure_kpa"].fillna(XD["final_hydrostatic_pressure_kpa"].median())
XD["maximum_shut-in_pressure_kpa"]= XD["maximum_shut-in_pressure_kpa"].fillna(XD["maximum_shut-in_pressure_kpa"].median())
```

```python
XD.info()
```

```python
XD.describe()
```

```python
yD = DuvTrueTempWellID["true_temperature_oc"]
```

```python
yD
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error as MSE
```

```python
XD_train, XD_test, yD_train, yD_test = train_test_split(XD,yD, test_size=0.3)
```

```python
rfD = RandomForestRegressor()
```

```python
params_rfD={'n_estimators':[50,100,200,300,500], 'max_features':['log2','auto','sqrt'],
            'min_samples_leaf':[1,2,5,10]}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid_rfD = GridSearchCV(estimator=rfD, param_grid=params_rfD, scoring ='neg_mean_squared_error', cv=3, verbose=1)
```

```python
grid_rfD.fit(XD_train, yD_train)
```

```python
best_modelD = grid_rfD.best_estimator_
```

```python
print(best_modelD)
```

```python
yD_pred= best_modelD.predict(XD_test)
```

```python
Drmse_test = MSE(yD_test, yD_pred)**(1/2)
print(Drmse_test)
```

```python
importancesD = pd.Series(data=best_modelD.feature_importances_,index=XD_train.columns)
importancesD_sorted = importancesD.sort_values()
```

```python
importancesD_sorted.plot(kind='barh', color = 'lightblue')
plt.title ('Features Importances')
plt.show()
```
